### Змінні
У Python змінні використовуються для зберігання даних, які можуть змінюватися під час виконання програми.

Основні типи даних в Python включають цілі числа (`int`), дійсні числа (`float`), рядки (`str`) та булеві значення (`bool`).

### Операції
Python підтримує різноманітні математичні та логічні операції.

```python
# Математичні операції
a = 10
b = 3
sum = a + b  # Додавання
difference = a - b  # Віднімання
product = a * b  # Множення
division = a / b  # Ділення
remainder = a % b  # Остача від ділення
power = a ** b  # Піднесення до степеня

print("Сума:", sum)
print("Різниця:", difference)
print("Добуток:", product)
print("Частка:", division)
print("Остача:", remainder)
print("Степінь:", power)

# Логічні операції
x = True
y = False
print(x and y)  # Логічне "І"
print(x or y)  # Логічне "АБО"
print(not x)  # Логічне "НЕ"
```

Ці приклади слугуватимуть хорошим стартом для освоєння основ Python. Важливо регулярно практикуватись та експериментувати з кодом, щоб краще зрозуміти мову програмування.

Логічні вирази в Python використовують логічні оператори (`and`, `or`, `not`) для формування виразів, які оцінюються як істинні (`True`) або хибні (`False`). Ці вирази часто застосовуються в умовних конструкціях (`if`, `elif`, `else`), а також в циклах та інших структурах керування потоком програми. Ось основні концепції логічних виразів у Python:

### Основні логічні оператори:

- **`and`** - повертає `True`, якщо обидва операнди істинні.
- **`or`** - повертає `True`, якщо хоча б один з операндів істинний.
- **`not`** - інвертує логічне значення операнда.

### Приклади логічних виразів:

```python
x = 5
print(x > 3 and x < 10)  # Виведе True, оскільки 5 більше 3 і одночасно менше 10

print(x > 3 or x < 4)  # Виведе True, оскільки одна з умов істинна (5 більше 3)

print(not(x > 3 and x < 10))  # Виведе False, оскільки вираз в дужках істинний, а оператор not інвертує його
```

### Використання в умовних конструкціях:

```python
age = 20
if age >= 18:
    print("Дорослий")
else:
    print("Неповнолітній")

is_student = False
if not is_student:
    print("Не студент")
```

### Логічні вирази з порівняннями:

Логічні вирази можуть включати порівняння, такі як `==` (рівність), `!=` (нерівність), `>` (більше), `<` (менше), `>=` (більше або рівно), `<=` (менше або рівно).

```python
a = 10
b = 20
c = 30

# Перевірка на діапазон
if a > 5 and b < 25:
    print("a більше 5 і b менше 25")

# Комбінація логічних операторів
if not (a > 15 or b < 15):
    print("a не більше 15 і b не менше 15")

# Використання логічних виразів для змінних різних типів
name = "Alice"
age = 22
if name == "Alice" or age < 30:
    print("Ім'я Alice або вік менше 30")
```

Логічні вирази є фундаментальними в Python і забезпечують потужний засіб для контролю потоку програми та реалізації складної логіки.

Для навчання контролю потоку в Python, особливу увагу слід приділити вивченню умовних операторів та циклів. Ось матеріали для самостійного вивчення з прикладами коду та коментарями.

### Умовні оператори (`if`, `elif`, `else`)
Умовні оператори дозволяють виконувати певний код залежно від того, чи виконується певна умова.

```python
# Приклад умовних операторів
age = 18

if age >= 18:
    print("Ви досягли повноліття.")
elif age < 0:
    print("Некоректний вік.")
else:
    print("Ви ще неповнолітні.")

# Вкладені умови
if age > 0:
    if age < 18:
        print("Ви ще неповнолітні.")
    else:
        print("Ви досягли повноліття.")
else:
    print("Некоректний вік.")
```

### Цикли (`for`, `while`)
Цикли використовуються для повторення виконання блоку коду декілька разів.

#### Цикл `for`
```python
# Проходження по діапазону чисел
for i in range(5):  # Виведе числа від 0 до 4
    print(i)

# Проходження по елементах списку
names = ["Анна", "Борис", "Віктор"]
for name in names:
    print(name)
```

#### Цикл `while`
```python
# Цикл while виконується, поки умова є істинною
count = 0
while count < 5:
    print(count)
    count += 1  # Важливо змінювати змінну, що контролює цикл
```

Ці приклади допоможуть вам зрозуміти, як контролювати потік виконання програми в Python за допомогою умовних операторів та циклів. Практикуйте ці концепції, створюючи власні програми, що використовують різні логічні умови та повторювані дії.

Функції в Python — це блоки коду, які викликаються за іменем, можуть приймати аргументи та повертати результат. Вони допомагають організувати код, роблячи його більш читабельним, ефективним та легким для повторного використання. Область видимості змінних визначає, де змінні можуть бути доступні в коді та змінювані.

### Створення та використання функцій

#### Оголошення функції
```python
def greet(name):
    """Виводить привітання."""
    print(f"Привіт, {name}!")
```

#### Виклик функції
```python
greet("Анна")  # Виведе: Привіт, Анна!
```

### Параметри та аргументи
Функції можуть приймати параметри, які є змінними в оголошенні функції, та аргументи, які передаються функції при її виклику.

```python
def power(base, exponent):  # base та exponent є параметрами
    return base ** exponent

result = power(2, 3)  # 2 та 3 є аргументами
print(result)  # Виведе: 8
```

### Значення за замовчуванням для параметрів
Функції можуть мати параметри зі значеннями за замовчуванням, які використовуються, якщо аргумент не переданий.

```python
def greet(name, greeting="Привіт"):
    print(f"{greeting}, {name}!")

greet("Анна")  # Використовує значення за замовчуванням для greeting
greet("Борис", "Добрий день")  # Передає аргумент для greeting
```

### Область видимості змінних
Змінні, оголошені всередині функції, мають локальну область видимості та не доступні за межами цієї функції. Змінні, оголошені за межами всіх функцій, мають глобальну область видимості та доступні будь-де в коді.

```python
x = "глобальна x"  # Глобальна змінна

def test():
    y = "локальна y"  # Локальна змінна
    print(y)
    print(x)  # Доступ до глобальної змінної

test()
# print(y)  # Це спричинить помилку, оскільки y є локальною змінною в функції
```

- **Уникнення дублювання коду**: Виділення спільного коду в функції.
- **Покращення читабельності**: Логічне розділення коду на блоки.

Колекції в Python — це вбудовані типи даних, які дозволяють зберігати набори інших об'єктів. Розглянемо чотири основні типи колекцій: списки, кортежі, словники та множини, із прикладами коду та коментарями для кращого розуміння.

### Списки (`list`)
Списки — це змінювані колекції, які дозволяють зберігати елементи в певному порядку.

```python
# Створення списку
fruits = ["яблуко", "банан", "вишня"]
print(fruits)

# Додавання елементу
fruits.append("апельсин")
print(fruits)

# Видалення елементу
fruits.remove("банан")
print(fruits)

# Доступ до елементу за індексом
print(fruits[1])  # вишня
```

### Кортежі (`tuple`)
Кортежі схожі на списки, але є незмінюваними. Вони можуть зберігати набір елементів в певному порядку.

```python
# Створення кортежу
dimensions = (200, 50)
print(dimensions)

# Спроба змінити елемент кортежу призведе до помилки
# dimensions[0] = 250  # TypeError

# Доступ до елементу кортежу
print(dimensions[0])
```

### Словники (`dict`)
Словники зберігають пари ключ-значення і дозволяють швидко знаходити значення за ключем.

```python
# Створення словника
car = {
    "марка": "Ford",
    "модель": "Mustang",
    "рік": 1964
}
print(car)

# Додавання або зміна елементу
car["колір"] = "червоний"
print(car)

# Доступ до значення за ключем
print(car["модель"])

# Видалення пари ключ-значення
del car["рік"]
print(car)
```

### Множини (`set`)
Множини — це незмінювані колекції, які не дозволяють дублікатів.

```python
# Створення множини
languages = {"Python", "Java", "C++"}
print(languages)

# Додавання елементу
languages.add("Python")  # Не додасться, бо вже є в множині
languages.add("Ruby")
print(languages)

# Видалення елементу (може викликати KeyError, якщо елементу немає)
languages.remove("Java")
print(languages)
```

Кожен тип колекції має свої особливості та призначення. Використання правильного типу колекції для конкретної задачі може значно підвищити ефективність та читабельність вашого коду. Практика роботи з цими колекціями допоможе вам краще зрозуміти, як організувати та маніпулювати даними в Python.

### Операції з файлами

#### Читання з файлу
Python дозволяє легко читати з файлів за допомогою вбудованих функцій. Важливо закрити файл після завершення роботи з ним, що можна зробити автоматично, використовуючи конструкцію `with`.

```python
# Читання з файлу
filename = 'example.txt'
with open(filename, 'r') as file:
    content = file.read()
    print(content)

# Читання файлу по рядках
with open(filename, 'r') as file:
    for line in file:
        print(line.strip())  # .strip() видаляє символи нового рядка
```

#### Запис у файл
Запис даних у файл також відбувається через вбудовані функції. Можна використовувати режим `'w'` для запису (що очищує файл перед записом) або `'a'` для додавання (що зберігає вміст файлу та додає нові дані в кінець).

```python
# Запис у файл
new_content = "Привіт, світ!"
with open('newfile.txt', 'w') as file:
    file.write(new_content)

# Додавання до файлу
with open('newfile.txt', 'a') as file:
    file.write("\nДодатковий рядок.")
```

### Обробка винятків

Використання `try` і `except` дозволяє обробляти помилки, які можуть виникнути під час виконання програми, не припиняючи її роботу.

```python
# Простий приклад обробки винятку
try:
    # спроба відкрити неіснуючий файл
    with open('does_not_exist.txt', 'r') as file:
        print(file.read())
except FileNotFoundError:
    print("Файл не знайдено.")

# Обробка декількох винятків
try:
    # код, який може спричинити різні винятки
    # наприклад, ділення на нуль або доступ до неіснуючого індексу списку
    result = 10 / 0
except ZeroDivisionError:
    print("Ділення на нуль!")
except IndexError:
    print("Індекс не знайдено!")
```

Використання цих конструкцій дозволяє вашому коду елегантно обробляти помилки та продовжувати виконання, замість негайного переривання при виникненні проблем. Практика роботи з файлами та обробка винятків є важливими для розробки надійних та стійких до помилок програм.

Об'єктно-орієнтоване програмування (ООП) є фундаментальним підходом у Python, який дозволяє структурувати програмне забезпечення як набір об'єктів, що взаємодіють між собою. ООП базується на трьох основних принципах: інкапсуляції, наслідуванні, та поліморфізмі. Далі ми зосередимося на класах, об'єктах, та наслідуванні.

### Класи та Об'єкти
Класи служать як шаблони для створення об'єктів, які є екземплярами цих класів. Об'єкти містять атрибути (дані) та методи (функції), які визначають їх поведінку.

#### Створення Класу та Об'єкта
```python
class Dog:
    # Ініціалізатор класу
    def __init__(self, name, age):
        self.name = name  # Атрибут інстанції
        self.age = age

    # Метод класу
    def bark(self):
        print(f"{self.name} каже Гав!")

# Створення об'єкта класу Dog
my_dog = Dog("Рекс", 2)
my_dog.bark()  # Виведе: Рекс каже Гав!
```

### Наслідування
Наслідування дозволяє одному класу (дочірньому класу) успадковувати атрибути та методи іншого класу (батьківського класу), що сприяє перевикористанню коду.

#### Приклад Наслідування
```python
# Батьківський клас
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Підклас повинен імплементувати абстрактний метод")

# Дочірній клас, який наслідує Animal
class Cat(Animal):
    def speak(self):
        return f"{self.name} каже Мяу!"

# Створення об'єкта дочірнього класу
my_cat = Cat("Васька")
print(my_cat.speak())  # Виведе: Васька каже Мяу!
```

### Поліморфізм
Поліморфізм — це здатність об'єктів різних класів відповідати на однакові повідомлення (тобто виклики методів) по-різному. Вищенаведений приклад з методом `speak` демонструє поліморфізм: метод `speak` викликається як для `Animal`, так і для `Cat`, але поведінка кожного об'єкта при цьому відрізняється.



Регулярні вирази в Python — потужний інструмент для пошуку, заміни та обробки текстових даних. Вони дозволяють визначати шаблони для пошуку у тексті, що робить можливим ефективну роботу з великими обсягами тексту або складними форматами даних.

### Основи регулярних виразів

Для роботи з регулярними виразами в Python використовується модуль `re`. Ось декілька основних функцій цього модуля:

- `re.search()`: Пошук першого збігу шаблону у рядку.
- `re.findall()`: Знаходження всіх збігів шаблону у рядку.
- `re.sub()`: Заміна знайдених збігів на інший рядок.

#### Приклади

##### Пошук у тексті
```python
import re

text = "Привіт, світ! Сьогодні 2024 рік."
# Пошук будь-яких цифр
match = re.search(r'\d+', text)
if match:
    print("Знайдено число:", match.group())  # Виведе: Знайдено число: 2024
else:
    print("Число не знайдено.")
```

##### Знаходження всіх збігів
```python
# Припустимо, ми шукаємо всі слова на "с"
words = re.findall(r'\bс\w+', text, re.IGNORECASE)
print("Знайдені слова:", words)  # Виведе список слів, що починаються на "с"
```

##### Заміна у тексті
```python
# Заміна всіх цифр на слово "рік"
modified_text = re.sub(r'\d+', 'рік', text)
print(modified_text)  # Привіт, світ! Сьогодні рік рік.
```

### Шаблони регулярних виразів

Регулярні вирази використовують спеціальну мову шаблонів для опису пошукових запитів. Ось деякі елементи цієї мови:

- `\d` - відповідає будь-якій цифрі.
- `\w` - відповідає будь-якому словесному символу (цифри, літери, підкреслення).
- `\s` - відповідає будь-якому пробільному символу (пробіл, табуляція).
- `.` - відповідає будь-якому символу, крім нового рядка.
- `+` - вказує, що попередній символ або група може зустрічатися один або більше разів.
- `*` - вказує, що попередній символ або група може зустрічатися нуль або більше разів.
- `?` - робить попередній символ або групу необов'язковими.

Вивчення та використання регулярних виразів може значно спростити обробку та аналіз текстових даних у ваших програмах

Цей код створює базовий веб-додаток, який відповідає на запити до головної сторінки (`/`) з текстом "Привіт, світ!".

### Автоматизація адміністративних завдань
Python може бути використаний для автоматизації різноманітних адміністративних завдань, таких як робота з базами даних, автоматичне відправлення електронних листів або управління серверами.

#### Приклад: Відправлення електронного листа
Для відправлення електронних листів можна використовувати модуль `smtplib`:
```python
import smtplib
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart

sender_email = "your_email@example.com"
receiver_email = "receiver@example.com"
password = input("Введіть пароль: ")

message = MIMEMultipart("alternative")
message["Subject"] = "Тестовий лист Python"
message["From"] = sender_email
message["To"] = receiver_email

# Створення текстової та HTML версій листа
text = """\
Привіт,
Це тестовий лист відправлений з Python."""
html = """\
<html>
  <body>
    <p>Привіт,<br>
       Це тестовий <b>лист</b> відправлений з Python.</p>
  </body>
</html>
"""

# Додавання текстової та HTML версій до MIMEMultipart об'єкта
part1 = MIMEText(text, "plain")
part2 = MIMEText(html, "html")

message.attach(part1)
message.attach(part2)

# Відправлення листа
with smtplib.SMTP_SSL('smtp.example.com', 465) as server:
    server.login(sender_email, password)
    server.sendmail(sender_email, receiver_email, message.as_string())
```

### 7. Розробка проектів
- **Практичні проекти**: Розробка власних проектів для закріплення навичок (to do list).



Розробка власного проекту "Список завдань" (To-Do List) є чудовим способом закріпити навички програмування на Python та ознайомитися з розробкою веб-додатків, роботою з базами даних, взаємодією з користувачем через веб-інтерфейс і, можливо, розгортанням додатку в інтернеті. Ось базові кроки для створення простого веб-додатку "Список завдань":

### Крок 1: Вибір інструментів
- **Backend**: Flask або Django для створення веб-додатку.
- **Frontend**: HTML, CSS (можливо, з Bootstrap для спрощення стилізації) та JavaScript для додавання динамічності.
- **База даних**: SQLite (для простоти) або PostgreSQL, якщо ви шукаєте більшу масштабованість.
- **ORM (Object-Relational Mapping)**: SQLAlchemy (для Flask) або Django ORM, щоб спростити роботу з базою даних через об'єктно-орієнтовані вирази.

### Крок 2: Створення базової структури проекту
- **Налаштування веб-сервера**: Створіть основний Flask або Django проект з одним додатком.
- **Моделі даних**: Визначте моделі для вашої бази даних (наприклад, `Task` з полями `id`, `title`, `description`, `completed`).
- **Маршрути та відображення (Views)**: Створіть маршрути для обробки запитів на додавання, видалення, та оновлення завдань.

### Крок 3: Розробка інтерфейсу користувача
- **HTML-шаблони**: Створіть шаблони для відображення списку завдань, форми додавання нових завдань, та кнопок для видалення та означення завдань як виконаних.
- **CSS/Bootstrap**: Стилізуйте ваші шаблони для покращення зовнішнього вигляду.
- **JavaScript**: Додайте фронтенд логіку для асинхронних запитів до сервера, якщо потрібно динамічне оновлення списку завдань без перезавантаження сторінки.

### Крок 4: Реалізація логіки додатку
- **CRUD операції**: Реалізуйте функціонал для створення, читання, оновлення та видалення завдань у вашій базі даних.
- **Взаємодія з користувачем**: Забезпечте користувачам можливість додавати нові завдання, відзначати завдання як виконані, та видаляти завдання.

### Крок 5: Тестування та розгортання
- **Тестування**: Перевірте ваш додаток на наявність помилок або недолік

ів у функціоналі.
- **Розгортання**: Розгорніть ваш додаток на публічному сервері за допомогою Heroku, DigitalOcean або іншого хостингу.

Створення проекту "Список завдань" дозволить вам застосувати знання Python та веб-розробки в практиці, розуміння принципів роботи веб-додатків, а також навички роботи з базами даних і фронтенд-технологіями.
